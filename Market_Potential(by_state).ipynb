```python
df1["Percent_Electric"] = (df1["Total Electric Vehicle"]/df1["Total"])*100
```

```python
df1["Non-Electric to Electric Ratio"] = df1["Total Non-Electric Vehicle"]/df1["Total Electric Vehicle"]
```

```python
df1.head(2)
```

```python
plt.figure(figsize=(45,20))
sns.barplot(x=df1["State Name"], y = df1["Percent_Electric"])
```

```python
plt.figure(figsize=(45,20))
plt.pie(df1["Non-Electric to Electric Ratio"], labels=df1["State Name"], autopct='%1.1f%%', startangle=90)
plt.title('Percent Non Electric')
```
