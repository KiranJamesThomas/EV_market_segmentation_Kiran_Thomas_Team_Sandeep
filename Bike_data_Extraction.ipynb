```python
import time
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver

# Set up Selenium WebDriver
driver = webdriver.Chrome()  # You need to have ChromeDriver installed. You can also use other drivers like Firefox or Safari.
url = "https://www.zigwheels.com/newbikes/best-electric-electric-two-wheelers-bikes"
driver.get(url)

# Scroll down the page to load all items
last_height = driver.execute_script("return document.body.scrollHeight")
while True:
    # Scroll down to bottom
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # Wait to load page
    time.sleep(2)

    # Calculate new scroll height and compare with last scroll height
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height


driver.execute_script("document.querySelector('.zw-cmn-loadMore').click();")
# Wait for some time for the additional items to load
time.sleep(2)

nu=30
while nu>0:
    driver.execute_script("document.querySelector('.zw-cmn-loadMore').click();")
    nu=nu-1
    time.sleep(2)

# Parse the HTML content of the webpage
soup = BeautifulSoup(driver.page_source, "html.parser")

# Find all the bike elements
bike_elements = soup.find_all("div", class_="p-10 pt-0 mt-negative")

# Initialize lists to store data
bike_data = []

# Extract bike names, prices, and range
for bike in bike_elements:
    name = bike.find("strong", class_="lnk-hvr block of-hid h-height").text.strip()
    price = bike.find("div", class_="clr-bl p-5").text.strip()
    range = bike.find("div", class_="clr-try fnt-14 pb-10 h-height of-hid").text.strip()

    # Append data to the list
    bike_data.append({"Bike Name": name, "Price": price, "Range": range})

# Create a DataFrame
df = pd.DataFrame(bike_data)

# Export data to CSV
df.to_csv("electric_bikes.csv", index=False)

print("Data extracted and saved to electric_bikes.csv")

```

```python
df1=pd.read_csv('electric_bikes.csv')

df1[['Kwh', 'Range_per_charge', 'Max_speed']] = df['Range'].str.split('|', expand=True)

# Remove leading and trailing whitespaces
df1['Kwh'] = df['Kwh'].str.strip()
df1['Range_per_charge'] = df['Range_per_charge'].str.strip()
df1['Max_speed'] = df['Max_speed'].str.strip()
df1.drop(columns=['Range'])
```
